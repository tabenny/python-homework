```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

#create a class to store the menu item info for each row
class MenuItem:
    def __init__(self, item, category,description,price,cost):
        self.item=item
        self.category=category
        self.description=description
        self.price=price
        self.cost=cost

        #this is the class for storing sales item
class SalesItem:
    def __init__(self, line_item_id, date,credit_card_number,quantity,menu_item):
        self.line_item_id=line_item_id
        self.date=date
        self.credit_card_number = credit_card_number
        self.quantity=quantity
        self.menu_item=menu_item
        
        
# @TODO: Read in the menu data into the menu list
with open(menu_filepath) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')

      # now first row is header, so skip that
    header = next(csv_reader)

    #print(header)
    
    for current_row in csv_reader:
        #initilaize a menu item and store the object to list
        tempMenuItem = MenuItem(current_row[0],current_row[1],current_row[2],float(current_row[3]),float(current_row[4]))
        menu.append(tempMenuItem)
                                
# @TODO: Read in the sales data into the sales list

with open(sales_filepath) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')

      # now first row is header, so skip that
    header = next(csv_reader)
    
    for current_row in csv_reader:
        tempSalesItem = SalesItem(int(current_row[0]),current_row[1],current_row[2],int(current_row[3]),current_row[4])
        sales.append(tempSalesItem)








# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every row in the sales list object


for curSalesItem in sales:
    quantity = curSalesItem.quantity
    sales_item = curSalesItem.menu_item

   
    #check whether the sales item is already present in the report dictionary, if its not add it
    #if its present, the following step will be skipped
    if sales_item not in report:
        tmpReportItem = {"01-count": 0,"02-revenue": 0,"03-cogs": 0,"04-profit": 0,}
        report.update({sales_item:tmpReportItem})
        
        
       # print(sales_item)
       # print(quantity)
       # print(report)
        
    #this variable is to check whether the item is found    
    bFound=False
    
    for curMenuItem in menu:
        item=curMenuItem.item
        price=curMenuItem.price
        cost=curMenuItem.cost
        #calculate the profit for each item
        profit=price-cost
        if(item==sales_item):
            bFound=True
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity
            break

    if bFound==False:
        print(f"("+ sales_item +") could not be found in the Menu")
#print(report)

#now loop through the report data and print
#now create the file to write to file
outF = open("sales_data_summary.txt", "w")

for curKey,curValue in report.items():
    print(curKey+ " "+str(curValue))
    outF.write(curKey+ " "+str(curValue))
    outF.write("\n")

outF.close()
```
